```python
web = BraneWeb()

web.add_seven_brane('7b_2', position=(-1, 0))
web.add_seven_brane('7b_3', position=(1, 0))

web.add_seven_brane('7b_bottom2', position=(1, -1))

web.add_seven_brane('7b_top1', position=(0, 1))

web.add_junction('j1', position=(0, 0))

web.add_brane('7b_2', 'j1', multiplicity=2)
web.add_brane('j1', '7b_3', multiplicity=1)

web.add_brane('7b_bottom2', 'j1', multiplicity=1)

web.add_brane('7b_top1', 'j1', multiplicity=1)

web.draw()
```

```python
edges1 = [('7b_2', 'j1', 0), ('j1', '7b_top1', 0), ('j1', '7b_bottom2', 0)]
edges2 = [('7b_2', 'j1', 1), ('j1', '7b_3', 0)]

subgraph1 = web.web.edge_subgraph(edges1)
subgraph2 = web.web.edge_subgraph(edges2)

web1 = BraneWeb.from_graph(subgraph1)
web2 = BraneWeb.from_graph(subgraph2)

web1.draw()
web2.draw()
```

```python
quivers = web.find_magnetic_quivers()

for i, quiv in enumerate(quivers):
    print(f"Magnetic quiver {i+1}:")
    quiv.draw()
```
