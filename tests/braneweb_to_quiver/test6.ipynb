```python
web = BraneWeb()

web.add_seven_brane('7b_0', position=(-3, 0))
web.add_seven_brane('7b_1', position=(-2, 0))
web.add_seven_brane('7b_2', position=(-1, 0))
web.add_seven_brane('7b_3', position=(1, 0))
web.add_seven_brane('7b_4', position=(2, 0))
web.add_seven_brane('7b_5', position=(3, 0))

web.add_seven_brane('7b_bottom1', position=(0, -1))
web.add_seven_brane('7b_bottom2', position=(1, -1))

web.add_seven_brane('7b_top1', position=(0, 1))
web.add_seven_brane('7b_top2', position=(-1, 1))

web.add_junction('j1', position=(0, 0))

web.add_brane('7b_0', '7b_1', multiplicity=1)
web.add_brane('7b_1', '7b_2', multiplicity=2)
web.add_brane('7b_2', 'j1', multiplicity=3)
web.add_brane('j1', '7b_3', multiplicity=3)
web.add_brane('7b_3', '7b_4', multiplicity=2)
web.add_brane('7b_4', '7b_5', multiplicity=1)

web.add_brane('7b_bottom1', 'j1', multiplicity=1)
web.add_brane('7b_bottom2', 'j1', multiplicity=1)

web.add_brane('7b_top1', 'j1', multiplicity=1)
web.add_brane('7b_top2', 'j1', multiplicity=1)

web.draw()
```

# Minimal subwebs across junction

```python
# checking minimal subwebs across junction j1
subwebs = web.find_subwebs_across_junction('j1')

for subweb in subwebs:
    subweb.draw()
```

# Decomposition

```python
subweb_decompositions = web.find_subweb_decompositions('j1', draw_subwebs=False)
print(len(subweb_decompositions), "decompositions found")
```
