```python
%cd ../../../src/

from unitary_quiver import Quiver 
from hasse import HasseDiagram
```

```python
number_of_nodes = 4

nodes = [i+1 for i in range(number_of_nodes)]
edges = [(1,2), (2,3), (3,4), (4,2)]
values = [1,4,4,4]

quiver = Quiver(nodes, edges, values)
quiver.draw()
```

```python
hasse = HasseDiagram.generate_from_quiver(quiver, image_path='../tests/unitary/repeated_subtraction')
hasse.plot_html('../tests/unitary/repeated_subtraction/repeated_subtraction.html')
```
