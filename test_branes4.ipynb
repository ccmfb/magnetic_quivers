```python
from brane_webs import Brane, BraneWeb, SuperBraneWeb, D7Brane
delta = 0.1
```

```python
D7s = [D7Brane() for _ in range(4+6+3+10)]
```

```python
webs = [
    # first 4 D7s: +1
    BraneWeb([
        Brane([0,0], [1,0], 1, starts_on=D7s[0], starts_on_side='inside', ends_on=D7s[1], ends_on_side='outside'),
    ]),
    BraneWeb([
        Brane([1,0], [2,0], 1, starts_on=D7s[1], starts_on_side='inside', ends_on=D7s[2], ends_on_side='outside'),
    ]),
    BraneWeb([
        Brane([2,0], [3,0], 1, starts_on=D7s[2], starts_on_side='inside', ends_on=D7s[3], ends_on_side='outside'),
    ]),
    BraneWeb([
        Brane([3,0], [4,0], 4, starts_on=D7s[3], starts_on_side='inside', ends_on=D7s[4], ends_on_side='inside'),
    ]),

    # next 6 D7s: +2
    BraneWeb([
        Brane([7,0], [6,0], 1, starts_on=D7s[4], starts_on_side='outside', ends_on=D7s[5], ends_on_side='inside'),
    ]),
    BraneWeb([
        Brane([8,0], [8,0], 1, starts_on=D7s[5], starts_on_side='outside', ends_on=D7s[6], ends_on_side='inside'),
    ]),
    BraneWeb([
        Brane([9,0], [10,0], 1, starts_on=D7s[6], starts_on_side='outside', ends_on=D7s[7], ends_on_side='inside'),
    ]),
]

for i in range(9):
    webs.append(
        BraneWeb([
            Brane([10+i,0], [9-i,0], 1, starts_on=D7s[7+i], starts_on_side='outside', ends_on=D7s[8+i], ends_on_side='inside'),
        ])
    )

webs.extend([
    BraneWeb([
        Brane([4,delta], [3,0], 3, ends_on=D7s[4], ends_on_side='inside'),
        Brane([4,delta], [-2,1], 5, ends_on=D7s[17], ends_on_side='inside'),
        Brane([4,delta], [-1,-1], 5, ends_on=D7s[18], ends_on_side='inside'),
        Brane([7,delta], [2,0], 1, starts_on=D7s[4], starts_on_side='outside', ends_on=D7s[5], ends_on_side='inside'),
        Brane([8,delta], [1,0], 1, starts_on=D7s[5], starts_on_side='outside', ends_on=D7s[6], ends_on_side='inside'),
    ]),
    BraneWeb([
        Brane([5,2*delta], [3,0], 2, ends_on=D7s[4], ends_on_side='inside'),
        Brane([5,2*delta], [0,-1], 4, ends_on=D7s[19], ends_on_side='inside'),
        Brane([5,2*delta], [-3,1], 5, ends_on=D7s[20], ends_on_side='inside'),
        Brane([7,2*delta], [2,0], 1, starts_on=D7s[4], starts_on_side='outside', ends_on=D7s[5], ends_on_side='inside'),
        Brane([8,2*delta], [1,0], 1, starts_on=D7s[5], starts_on_side='outside', ends_on=D7s[6], ends_on_side='inside'),
    ]),
    BraneWeb([
        Brane([6,3*delta], [3,0], 1, ends_on=D7s[4], ends_on_side='inside'),
        Brane([6,3*delta], [-2,-1], 6, ends_on=D7s[21], ends_on_side='inside'),
        Brane([6,3*delta], [-1,1], 4, ends_on=D7s[22], ends_on_side='inside'),
        Brane([7,3*delta], [2,0], 1, starts_on=D7s[4], starts_on_side='outside', ends_on=D7s[5], ends_on_side='inside'),
        Brane([8,3*delta], [1,0], 1, starts_on=D7s[5], starts_on_side='outside', ends_on=D7s[6], ends_on_side='inside'),
    ])

])
```

```python
super_web = SuperBraneWeb(webs)

super_web.plot(figsize=(12, 6))

for i in range(len(super_web)):
    for j in range(len(super_web)):
        if i < j:
            edges = super_web.edges(i, j)

            if edges != 0:
                print(f'Edges between {i} and {j}: {edges}')
            
```
