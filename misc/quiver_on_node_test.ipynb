```python
from pyvis.network import Network

# Create a network
net = Network(
    height="600px", 
    width="100%", 
    bgcolor="#f0f0f0", 
    font_color="black"
)

# Example 1: simple formatting
html1 = """
<div style="padding:10px;">
  <h3 style="margin:0;">Node 1</h3>
  <p>This is a <b>bold</b> statement and an <i>italic</i> one.</p>
</div>
"""
net.add_node(
    1,
    label="Node 1",
    title=html1  # <--- your HTML goes here
)

# Example 2: image + link
html2 = """
<div style="text-align:center;">
  <img src="https://via.placeholder.com/80" 
       alt="placeholder" 
       style="display:block;margin:0 auto 8px;"/>
  <a href="https://example.com" target="_blank">
    Visit Example.com
  </a>
</div>
"""
net.add_node(
    2,
    label="Node 2",
    title=html2
)

# Connect them
net.add_edge(1, 2)

# Optional: show the control panel so you can tweak physics / layout in-browser
net.show_buttons(filter_=["physics"])

#net.show("pyvis_html_tooltips.html")
net.write_html("pyvis_html_tooltips.html")
```
