```python
from brane_webs import Brane, BraneWeb, SuperBraneWeb, D7Brane
delta = 0.1
```

```python
D7s = [D7Brane() for _ in range(5)]
```

```python
web1 = BraneWeb([
    Brane([0,0], [-1,-1], 1, ends_on=D7s[0], ends_on_side='inside'),
    Brane([0,0], [-1,1], 1, ends_on=D7s[1], ends_on_side='inside'),
    Brane([0,0], [2,0], 1, ends_on=D7s[2], ends_on_side='inside'),
    Brane([1,0], [1,0], 1, starts_on=D7s[2], starts_on_side='outside', ends_on=D7s[3], ends_on_side='inside'),
])

web2 = BraneWeb([
    Brane([-1,-delta], [1,0], 2, starts_on=D7s[4], starts_on_side='inside', ends_on=D7s[2], ends_on_side='inside'),
])
```

```python
super_web = SuperBraneWeb([web1, web2])

print('Intersection Number:', super_web.intersection_number(0,1))
super_web.plot()
```
