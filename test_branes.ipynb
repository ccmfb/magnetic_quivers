```python
from brane_webs import Brane, BraneWeb, SuperBraneWeb
delta = 0.1
```

```python
web1 = BraneWeb([
    Brane([0,0], [-1,0], 1),
    Brane([0,0], [0,1], 1),
    Brane([0,0], [1,-1], 1),
])

web2 = BraneWeb([
    Brane([-delta, delta], [1,0], 1),
    Brane([-delta, delta], [0,-1], 1),
    Brane([-delta, delta], [-1,1], 1),
])
```

```python
super_web = SuperBraneWeb([web1, web2])

print('Intersection Number:', super_web.intersection_number(0,1))
super_web.plot()
```
