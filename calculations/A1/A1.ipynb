```python
def generate_finite_quiver(n):
    number_nodes = 2 * n + 1

    nodes = [i+1 for i in range(number_nodes)]
    edges = [(i+1, i+2) for i in range(number_nodes - 3)] + [(n, number_nodes - 1), (n, number_nodes)]
    values = [i+1 for i in range(n)] + [n-i-1 for i in range(n-1)] + [1, 1]

    return Quiver(nodes, edges, values)


def generate_infinite_quiver(n):
    number_nodes = 2 * n + 2

    nodes = [i+1 for i in range(number_nodes)]
    edges = [(i+1, i+2) for i in range(number_nodes - 4)] + [(number_nodes - 2, number_nodes - 1), (number_nodes - 1, number_nodes), (number_nodes - 1, n)]
    values = [i+1 for i in range(n)] + [n-i-1 for i in range(n-1)] + [1, 2, 1]

    return Quiver(nodes, edges, values)
```

```python
f_quiver = generate_finite_quiver(8)
i_quiver = generate_infinite_quiver(8)
f_quiver.display()
i_quiver.display()
```

```python
f_hasse = HasseDiagram.generate_from_quiver(f_quiver)
i_hasse = HasseDiagram.generate_from_quiver(i_quiver)

f_hasse.plot_html('calculations/A1/finite.html')
i_hasse.plot_html('calculations/A1/infinite.html')
```
