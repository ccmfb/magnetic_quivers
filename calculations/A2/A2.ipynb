```python
%cd ../../src/

from unitary_quiver import Quiver 
from hasse import HasseDiagram
```

```python
def generate_finite_finite_quiver(n,r):
    number_nodes = 2 * n + 3 * r + 2

    nodes = [i+1 for i in range(number_nodes)]
    edges = [(i+1, i+2) for i in range(number_nodes - 4)] + [(n+r, number_nodes - 2), (n+r, number_nodes - 1), (n+r, number_nodes)]
    values = [i+1 for i in range(n)] + [n + 2*(i+1) for i in range(r)] + [n + 2*r - (i+1) for i in range(n + 2*r - 1)] + [1, 1, 1]

    return Quiver(nodes, edges, values)


def generate_finite_infinite_quiver(n,r):
    number_nodes = 2 * n + 3 * r + 3

    nodes = [i+1 for i in range(number_nodes)]
    edges = [(i+1, i+2) for i in range(number_nodes - 5)] + [(n+r, number_nodes - 3), (number_nodes - 2, number_nodes - 1), (number_nodes - 1, n+r), (number_nodes - 1, number_nodes)]
    values = [i+1 for i in range(n)] + [n + 2*(i+1) for i in range(r)] + [n + 2*r - (i+1) for i in range(n + 2*r - 1)] + [1, 1, 2, 1]

    return Quiver(nodes, edges, values)


def generate_infinite_infinite_quiver(n,r):
    number_nodes = 2 * n + 3 * r + 4

    nodes = [i+1 for i in range(number_nodes)]
    edges = [(i+1, i+2) for i in range(number_nodes - 6)] + [
        (number_nodes - 4, number_nodes - 3),
        (number_nodes - 3, number_nodes - 2),
        (number_nodes - 2, number_nodes - 1),
        (number_nodes - 1, number_nodes),
        (n+r, number_nodes - 2),
    ]
    values = [i+1 for i in range(n)] + [n + 2*(i+1) for i in range(r)] + [n + 2*r - (i+1) for i in range(n + 2*r - 1)] + [1, 2, 3, 2, 1]

    return Quiver(nodes, edges, values)
```

Different phases for n = 3:

```python
n = 4
r = 2

ff_quiver = generate_finite_finite_quiver(n,r)
fi_quiver = generate_finite_infinite_quiver(n,r)
ii_quiver = generate_infinite_infinite_quiver(n,r)

ff_quiver.display()
fi_quiver.display()
ii_quiver.display()
```

```python
ff_hasse = HasseDiagram.generate_from_quiver(ff_quiver, image_path='../calculations/A2')
ff_hasse.plot_html('../calculations/A2/ff_hasse.html')
```

```python
fi_hasse = HasseDiagram.generate_from_quiver(fi_quiver, image_path='../calculations/A2')
fi_hasse.plot_html('../calculations/A2/fi_hasse.html')
```
