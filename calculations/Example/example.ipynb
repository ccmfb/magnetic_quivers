```python
%cd ../../src

from unitary_quiver import Quiver
from hasse import HasseDiagram
```

```python
n = 6

number_nodes = 2 * n + 1
nodes = [i+1 for i in range(number_nodes)]
edges = [(i+1, i+2) for i in range(number_nodes - 3)] + [(n, number_nodes - 1), (n, number_nodes)]
values = [i+1 for i in range(n)] + [n-i-1 for i in range(n-1)] + [1, 1]

quiver = Quiver(nodes, edges, values)
quiver.draw()
```

```python
hasse = HasseDiagram.generate_from_quiver(quiver, image_path='../calculations/Example')
hasse.plot_html('../calculations/Example/hasse.html')
```
