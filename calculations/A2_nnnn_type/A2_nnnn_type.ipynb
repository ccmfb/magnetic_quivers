```python
%cd ../../src/

from unitary_quiver import Quiver 
from hasse import HasseDiagram
```

```python
def generate_finite_finite_quiver(n):
    number_nodes = 2 * n + 2

    nodes = [i+1 for i in range(number_nodes)]
    edges = [(i+1, i+2) for i in range(number_nodes - 4)] + [(n, number_nodes - 2), (n, number_nodes - 1), (n, number_nodes)]
    values = [i+1 for i in range(n)] + [n-i-1 for i in range(n-1)] + [1, 1, 1]

    return Quiver(nodes, edges, values)

def generate_finite_infinite_quiver(n):
    number_nodes = 2 * n + 3

    nodes = [i+1 for i in range(number_nodes)]
    edges = [(i+1, i+2) for i in range(number_nodes - 5)] + [
        (n, number_nodes - 3), (n, number_nodes - 2), (number_nodes - 2, number_nodes - 1), (number_nodes - 2, number_nodes)
    ]
    values = [i+1 for i in range(n)] + [n-i-1 for i in range(n-1)] + [1, 2, 1, 1]

    return Quiver(nodes, edges, values)

def generate_infinite_infinite_quiver(n):
    number_nodes = 2 * n + 4

    nodes = [i+1 for i in range(number_nodes)]
    edges = [(i+1, i+2) for i in range(number_nodes - 6)] + [
        (number_nodes - 4, number_nodes - 3), (number_nodes - 3, number_nodes - 2), (number_nodes - 2, number_nodes - 1), 
        (number_nodes - 1, number_nodes), (number_nodes - 2, n)
    ]
    values = [i+1 for i in range(n)] + [n-i-1 for i in range(n-1)] + [1, 2, 3, 2, 1]

    return Quiver(nodes, edges, values)
```

Different phases for n = 3:

```python
n = 8

ff_quiver = generate_finite_finite_quiver(n)
fi_quiver = generate_finite_infinite_quiver(n)
ii_quiver = generate_infinite_infinite_quiver(n)

ff_quiver.display()
fi_quiver.display()
ii_quiver.display()
```

```python
ff_hasse = HasseDiagram.generate_from_quiver(ff_quiver)
fi_hasse = HasseDiagram.generate_from_quiver(fi_quiver)
ii_hasse = HasseDiagram.generate_from_quiver(ii_quiver)

ff_hasse.plot_html('../calculations/A2_nnnn_type/ff_hasse.html')
fi_hasse.plot_html('../calculations/A2_nnnn_type/fi_hasse.html')
ii_hasse.plot_html('../calculations/A2_nnnn_type/ii_hasse.html')
```
