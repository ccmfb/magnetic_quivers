```python
braneweb = BraneWeb()

braneweb.add_seven_brane('s1', position=(-2,0))
braneweb.add_seven_brane('s2', position=(-1,0))

braneweb.add_seven_brane('s3', position=(1,0))
braneweb.add_seven_brane('s4', position=(2,0))
braneweb.add_seven_brane('s5', position=(3,0))

braneweb.add_seven_brane('st1', position=(-2,1))
braneweb.add_seven_brane('st2', position=(-1,1))
braneweb.add_seven_brane('st3', position=(1,1))

braneweb.add_seven_brane('sb1', position=(-1,-1))
braneweb.add_seven_brane('sb2', position=(0,-1))
braneweb.add_seven_brane('sb3', position=(2,-1))

braneweb.add_junction('j1', position=(0,0))

braneweb.add_brane('s1', 's2', multiplicity=1)
braneweb.add_brane('s2', 'j1', multiplicity=2)

braneweb.add_brane('j1', 's3', multiplicity=3)
braneweb.add_brane('s3', 's4', multiplicity=2)
braneweb.add_brane('s4', 's5', multiplicity=1)

braneweb.add_brane('j1', 'st1', multiplicity=1)
braneweb.add_brane('j1', 'st2', multiplicity=1)
braneweb.add_brane('j1', 'st3', multiplicity=1)

braneweb.add_brane('j1', 'sb1', multiplicity=1)
braneweb.add_brane('j1', 'sb2', multiplicity=1)
braneweb.add_brane('j1', 'sb3', multiplicity=1)

braneweb.draw()
```

```python
subwebs = braneweb.subwebs()

for subweb in subwebs:
    webweb = BraneWeb.from_subgraph_edges(braneweb.web, subweb)
    webweb.draw()
```

```python
decompositions = braneweb.decompositions()
print(len(decompositions))

for decomposition in decompositions:
    print('Decomposition')
    for subweb in decomposition:
        webweb = BraneWeb.from_subgraph_edges(braneweb.web, subweb)
        webweb.draw()
```
